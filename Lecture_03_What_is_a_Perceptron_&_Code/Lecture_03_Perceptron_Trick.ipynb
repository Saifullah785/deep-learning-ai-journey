# **Perceptron Trick | How to train a Perceptron | Perceptron Part 2 |**

```python
# importing libraries

from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
```

```python
# This cell generates a synthetic dataset for binary classification.

x,y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                          n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
# This cell displays the target variable y.
y
```

```python
# This cell visualizes the generated dataset.


plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter', s=100)
plt.show()
```

```python
# This cell defines the perceptron training function

def perceptron(x,y):
  x = np.insert(x,0,1,axis=1)
  weights = np.ones(x.shape[1])
  lr = 0.1

  for i in range(1000):
    j= np.random.randint(0,100)
    y_hat = step(np.dot(x[j],weights))
    weights = weights + lr*(y[j]-y_hat)*x[j]
  return weights[0],weights[1:]
```

```python
# This cell defines the step function.

def step(z):
  return 1 if z >0 else 0
```

```python
# This cell trains the perceptron model and gets the intercept and coefficients.

intercept_,coef_ = perceptron(x,y)
```

```python
# This cell prints the learned coefficients and intercept.

print(coef_)
print(intercept_)
```

```python
# This cell calculates the slope (m) and y-intercept (b) of the decision boundary

m = - (coef_[0]/coef_[1])
b = - (intercept_/coef_[1])
```

```python
# This cell generates input values for plotting the decision boundary.

x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
# This cell plots the dataset and the learned decision boundary.

plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter', s=100)
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.ylim(-3,2)
```
