```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/DailyDelhiClimateTest.csv')
```

```python
df.head()
```

```python
df = df[['date','meantemp']]
```

```python
df.head()
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
```

```python
x1 =  df['meantemp'].ewm(alpha=0.9).mean()
```

```python
x1
```

```python
df['ewma'] = x1
df
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
plt.plot(df['date'],x1,df['ewma'],color='blue',label='alpha= 0.9')
plt.legend()
plt.show()
```
