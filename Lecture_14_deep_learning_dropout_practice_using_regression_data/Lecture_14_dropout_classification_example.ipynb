# Dropout model

```python
# Define a new sequential model for the neural network with Dropout layers
model = Sequential()

# Add the first dense layer with 128 units and ReLU activation. Input dimension is 2.
model.add(Dense(128 ,input_dim = 2 , activation = 'relu' ))
# Add a Dropout layer with a dropout rate of 0.25
model.add(Dropout(0.25))
# Add a second dense layer with 128 units and ReLU activation
model.add(Dense(128 , activation='relu'))
# Add another Dropout layer with a dropout rate of 0.25
model.add(Dropout(0.25))
# Add the output layer with 1 unit and sigmoid activation for binary classification
model.add(Dense(1, activation='sigmoid'))
```

```python
# Define the Adam optimizer with a learning rate of 0.01
adam = Adam(learning_rate=0.01)

# Compile the model with binary crossentropy loss, Adam optimizer, and accuracy metric
model.compile(loss='binary_crossentropy' , optimizer=adam , metrics=['accuracy'])
```

```python
# Train the model on the data with 100 epochs and a validation split of 0.2
history = model.fit(X, y, epochs=100, validation_split=0.2, verbose=1)
```

```python
# Plot the decision boundary of the trained model
plot_decision_regions(X,y.astype('int'), clf=model, legend=2)
# Set the x and y axis limits for the plot
plt.xlim(-0.7,0.5)
plt.ylim(-0.8,0.8)
# Show the plot
plt.show()
```

```python
# Plot the training and validation loss over epochs
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
# Plot the training and validation accuracy over epochs
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
